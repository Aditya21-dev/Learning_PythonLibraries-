```python
import numpy as np

a = np.array([[1,2,3,4],
              [5,6,7,8]])

print(np.ravel(a))

# return view
# consumes less memory
# faster
# original will be change if we change in view 
```

```python
import numpy as np

a = np.array([1,2,3,4])
b = np.array([2,4,5,7])

print(np.intersect1d(a,b,assume_unique=True,return_indices=True))
```

```python
import numpy as np
from functools import reduce


a = np.array([1,2,3,4])
b = np.array([2,4,5,7])
c = np.array([8,9,5,7])

print(reduce(np.union1d,(a,b,c)))
```
