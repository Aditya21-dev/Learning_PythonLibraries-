```python
import numpy as np

a = np.array([1,2,3,4,5])
b = np.copy(a)
c = a.view()

np.delete(b,4)
np.delete(c,4)

print(a)
print(b)
print(c)
```

```python
import numpy as np

a = np.array([1,2,3,4,5,6])
a2 = np.array([6,7,8,9,10])

print(np.setxor1d(a,a2))
print(np.setdiff1d(a,a2))
```

```python
import numpy as np

a = np.array([[1,2],
              [3,4]])
b = np.array([[5,6],
              [7,8]])

print(np.dot(a,b))
```

```python
import numpy as np

a = np.array([[1,2,13],
              [3,4,14],
              [11,12,15]])
b = np.array([[5,6,16],
              [7,8,17],
              [9,10,18]])

print(np.cross(a,b))
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])

print(np.add(a,b))
print(np.subtract(a,b))
print(np.divide(a,b))
print(np.multiply(a,b))
print(np.power(a,b))
print(np.sqrt(a))
```
