```python
import numpy as np
import time

n = 10

#lIST
list1 = list(range(n))
list2 = list(range(n))

start = time.time()
result = [list1[i] + list2[i] for i in range (n)]
list_time = time.time() - start

# numpya arry


ar1 = np.arange(n, dtype=np.int32)
ar2 = np.arange(n, dtype=np.int32)

start2 = time.time()
result_arr = ar1 + ar2 
numpy_time = time.time() - start2

print(f'List Time   : {list_time:.6f} sec')
print(f'Numpy Time  : {numpy_time:.6f} sec')
print(f'Numpy is approximetly {list_time/numpy_time:.2f}x faster')
```
