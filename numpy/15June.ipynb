```python
import numpy as np

arr = np.array([5, -2, 8, -1, 10, -7])

for i in range(len(arr)):
    if arr[i] < 0:
        arr[i] = 0

print(arr)
```

```python
import numpy as np

arr = np.array([3, 8, 12, 7, 15, 20])

count = 0

for i in arr:
    if i % 2 == 0:
        count += 1

print(count)
```

```python
import numpy as np

arr = np.array([1,2,3,2,4,5,1,6])

duplicates = []

for i in range(len(arr)):
    for j in range(i+1, len(arr)):
        if arr[i] == arr[j]:
            if arr[i] not in duplicates:
                duplicates.append(arr[i])

print(duplicates)
```
