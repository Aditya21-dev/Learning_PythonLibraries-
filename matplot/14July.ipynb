```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")

a = df.groupby("day")["tip"].mean()

a 

# plt.figure()
# plt.plot(a,)


plt.figure(figsize=(6,4),facecolor="green")

plt.plot(a.index, a.values, marker='o')

plt.title("Average Tip by Day")
plt.xlabel("Day")
plt.ylabel("Average Tip")
# plt.
plt.grid(True)

plt.show()
```

```python

a = df.groupby("day")["size"].sum()

a


plt.figure(figsize=(6,4))

plt.bar(a.index, a.values)

plt.title("Total customers by Days")
plt.xlabel("Daya")
plt.ylabel("no. of customers")

plt.show()
```

```python

# customers = df["size"].sum()

# smoker_percentage =  151*100/customers

# smoker_percentage



smoker_percentage = df["smoker"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(smoker_percentage.values , labels=smoker_percentage.index , autopct='%1.1f%%')
plt.title("smoker and non-smokers ")
plt.show()
```

```python
x = np.linspace(0,10,100)
fig , ax = plt.subplots(figsize=(5,3.2))
ax.plot(x , np.sin(x) , label='sin(x)' , color='#2b6cb0')

ax.plot(x , np.cos(x) , label = 'cos(x)' , color='#e53e3e' , linestyle='--')

ax.set_title('line plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```
