```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.__version__)
```

```python
df = sns.load_dataset('titanic')
df.head(5)
```

```python
import matplotlib.pyplot as plt

# Count passengers for each age
age_count = df['age'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.bar(age_count.index, age_count.values)

plt.title("Passenger Count by Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
# ages = df['age']
plt.figure(figsize=(5,4))
plt.hist(df['age'].dropna() , bins=25)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
count , bins = np.histogram(df["age"].dropna() , bins=25)

plt.figure(figsize=(5 , 4))
plt.stairs(count , bins , fill=True)


plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.savefig("First_plot")

plt.show()
```

```python
fig , ax  = plt.subplots(2 , 2 , figsize=(8,8))
ax[0,0].hist(df['age'] , bins = 20 , color = "skyblue")
ax[0,0].set_title("Age Distribution")

ax[0,1].hist(df['fare'],bins=20 , color = "orange")


class_counts =  df['class'].value_counts()
ax[1,0].bar(class_counts.index,class_counts.values,color="green")
ax[1,0].set_title("Pasangers by Class")

Survival = df.groupby("sex")['survived'].mean()
ax[1,1].bar(Survival.index,Survival.values,color="pink")
ax[1,1].set_title("Survival Rate by Gender")


plt.suptitle("Titanic Dashboard")
plt.tight_layout()
plt.show()
```

```python
cols = ['survived' , 'pclass' , 'age' , 'sibsp' , 'parch' , 'fare']

corr = df[cols].corr()

corr 

plt.figure()

sns.heatmap(corr , annot=True , cmap='coolwarm' , vmin= -1 , vmax= 1)
plt.title("Feature Correlation Map")
plt.show()
```
