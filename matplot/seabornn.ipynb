```python
df['age_group'] = pd.cut(df['age'], bins = range(0, 85, 10))

surv_by_age = df.groupby('age_group')['survived'].mean()


# plt.figure(figsize=(5,5))
surv_by_age.plot(kind = 'line', marker = 'o', figsize=(7,7))
plt.title("Survival rate by Age")
plt.xlabel("Age Group")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

# # Load Titanic dataset
# df = sns.load_dataset("titanic")

# # Remove rows where age or fare is missing
# df = df.dropna(subset=["age", "fare"])

# # Create scatter plot
# plt.figure(figsize=(8, 5))
# plt.scatter(df["age"], df["fare"], alpha=0.6)

# # Labels and title
# plt.title("Scatter Plot: Age vs Fare")
# plt.xlabel("Age")
# plt.ylabel("Fare")

# # Show plot
# plt.grid(True)
# plt.show()




import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")
df = df.dropna(subset=["age", "fare"])

plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x="age",
    y="fare",
    hue="survived",
    palette="Set1",
    alpha=0.7
)

plt.title("Age vs Fare (Colored by Survival)")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

```python
fig , ax = plt.subplots(figsize=(8,4))
sc = ax.scatter(df['age'] , df['fare'],
                c = df['survived'] , cmap='coolwarm')

ax.set_title('Scatter plot -- Age / Survived')
ax.set_xlabel('Age')
ax.set_ylabel('Fare (dollar)')
cbar = fig.colorbar(sc, ax=ax , ticks = [0,1])
cbar.ax.set_yticklabels(['Died' , 'Survived'])

plt.show()
```

```python
grp = df.groupby(['class' , 'sex'])['survived'].mean().unstack()

x = np.arange(len(grp.index))
w = 0.35

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.bar(x - w/2 , grp['male'] , w , label = "Male")
plt.bar(x + w/2 , grp['female'] , w , label = "Female")

plt.xticks(x , grp.index)
plt.title("survival Rate")
plt.legend()



stack = df.groupby(['class' , 'survived']).size().unstack() 

plt.subplot(1,2,2)

plt.bar(stack.index , stack[0] , label='Died')
plt.bar(stack.index , stack[1] , bottom=stack[0] , label = 'Survived')
plt.title("Outcome Counts")
plt.legend()

plt.tight_layout()
plt.show()
```
