```python
df['age'].dropna(inplace=True)
df
```

```python
df['age'] = df['age'].fillna(df.age.median())

df['age']
```

```python
# Q . findthe survival rate bn age group 

age_bins = np.arange(0,85,5)
df['age_group'] = pd.cut(df['age'] , bins=age_bins)
```

```python
df["age_group"] = df.groupby('age')['survived'].mean()
surv_by_age = df.groupby('age_group')['survived'].mean()

plt.figure()

surv_by_age.plot(surv_by_age.index,surv_by_age.values , marker = 'o')
plt.title("survival rate by Age")
plt.xlabel("age group")
plt.ylabel("survival rate")
plt.show()
```
