## DATA FRAME

```python
# Create data_frame from LIST

import pandas as pd

data = [["shivangi" , 19 ,"indore"],
        ["aditya" , 19 , "bhopal"],
        ["sandhya" , 20 , "simliguda"],
        ["ramesh" , 25 ,"indore"],
        ["mahipal" , 25 , "bhopal"]]

df = pd.DataFrame(data , columns=["Nmae" , "Age" , "City"])

print(df) 
```

```python
# Create data_frame from DICTIONARY 

data = {
"Name": ["Alice", "Bob", "Carol"],
"Age" : [25, 30, 22],
"City": ["Delhi", "Mumbai", "Pune"]
}
df = pd.DataFrame(data)
print(df)
```
