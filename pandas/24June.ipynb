## Mathamatical Functions

```python
import numpy as np
import pandas as pd
a = np.array([ 10 , 20 , np.nan , 30 , np.nan , 40 , 50 , 60 , np.nan , 70 , np.nan , 80])

se = pd.Series(a)

se

se.size 

se.count()

se.sum()

se.product()

se.mean()

se.mode()

se.std()

se.var()

se.min()

se.max()

se.describe()
```

## Null Values Handeling 

```python
import pandas as pd

a = np.array([ 10 , 20 , np.nan , 30 , np.nan , 40 , 50 , 60 , np.nan , 70 , np.nan , 80])
se = pd.Series(a)

se.isnull()
se.isnull().sum()
```

```python
se.notnull()
se.notnull().sum()
```

```python
se.hasnans
```
