##PYthon  IN Build Functions

```python
import pandas as pd
import numpy as np
a = np.array([ 10 , 20 , np.nan , 30 , 20 , 40 , 50 , np.nan , 40 , 50 , 60 , np.nan , 70 , np.nan , 80])
se = pd.Series(a)

se.unique()
```

```python
print(len(se))
print(max(se))
print(min(se))
print(type(se))

print(dir(se))
```

```python
print(sorted(se))
```

##Typpe Conversion 


```python
import numpy as np
import pandas as pd

a = [ 10 , 20 , 30 , 40]
se = pd.Series(a)

print(se.astype(int))

print(se.astype(float))

print(se.astype(bool))

print(se.astype(str))
```

```python
list(se)

tuple(se)

dict(se)

set(se)
```

#   Membership operator

```python
# check indexing is in or not 

print(0 in se)

print(5 in se) 
```

```python
#  check values is in or not 

print(se.isin([10,20,40]))
```

#   Looping over a SERIES

```python
for i in se:
    print(i)
```

```python
for i in se.keys():
    print(i)
```

```python
for i ,v in se.items():
    print(i,v)
```

```python


def square(n):
    return n * n
print(se.apply(square))

```
